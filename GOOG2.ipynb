# GOOG Performance Algo

```python
import pandas as pd
```

```python
df=pd.read_csv("~/Downloads/GOOG.csv")
```

```python
dates=list(df['Date'])
```

```python
prices=list(df['Close'])
```

```python
def get_daily_return(old_price,new_price):
  return (new_price-old_price)/old_price
```

```python
#Here is what prices is now, just a big list of prices
print(prices)
```

```python
# You can get a single price by doing what is called "indexing" a list.  So if I want the first price I can do
prices[0]
```

```python
#Second one... you can see python is a "zero indexed language", meaning the first item in an interable is a 0, not a 1.
prices[1]
```

```python
#You can also do math to get the index.  Here I want the third one
prices[1+1]

#
```

```python
#ok so now that you get indexing that prices list, look where it calls the function, get_daily_return(prices[i],prices[i+1])
# it is passing a variable 'i' to the indexer.  So lets add a print to so you can see what is going on.

returns_list=[]
for i in range(0,len(prices)-1):
    print ('Iteration: ' + str(i))
    print ('Old price: ')
    print (prices[i])
    print ('New Price: ')
    print (prices[i+1])
    output=get_daily_return(prices[i],prices[i+1])

    returns_list.append(output)
```

```python
def buy_sell(num):
    if num < 0:
        recommendation="Buy"
    else:
        recommendation="Sell"
    return recommendation
```

```python
returns_list=[]
recommendations_list=[]
for i in range(0,len(prices)-1):

    output=get_daily_return(prices[i],prices[i+1])
    recommendation=buy_sell(output)
    returns_list.append(output)
    recommendations_list.append(recommendation)
```

```python
returns_list.insert(0,0)
```

```python
recommendations_list.insert(0,"")
```

```python
summary_df=pd.DataFrame({'Date':dates,'Closing Price':prices,'Daily Return':returns_list,'Reco':recommendations_list})
```
